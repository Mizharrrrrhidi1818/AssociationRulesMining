```python
import re
import emoji
from googleapiclient.discovery import build

# Step 1: Initiation API YouTube API Key
API_KEY = 'AIzaSyBUug4PcM5edTDQUwHY1ldH_Z5vXg2zTOE'  # Put in Api Key

youtube = build('youtube', 'v3', developerKey=API_KEY)   # Inisialitation API Youtube

# Step 2: Take it YouTube account and Extract video ID
video_url = input ('Input URL Video youtube')
video_id = video_url [-11:]   # take 11 last character from URL as Video ID
print("ID Videoc :" + video_id)

# Step 3: Obtain Channel ID from Youtube Account
video_response = youtube.videos().list(
    part='snippet',
    id=video_id
).execute()

# Extract Channel ID from respons
video_snippet = video_response['items'][0]['snippet']
uploader_channel_id = video_snippet['channelId']
print("ID Channel : " + uploader_channel_id)

# Step 4: Take comments
print("Take comments...")
comments = []
nextPageToken = None
while len(comments) < 2000:
  request = youtube.commentThreads().list(
      part='snippet',
      videoId=video_id,
      maxResults=100, # Take 100 comments per-request
      pageToken = nextPageToken
  )
  response = request.execute()
  for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']

    # Investigate if comment is not from video uploader
    if comment['authorChannelId']['value'] != uploader_channel_id:
      comment_text = comment['textDisplay']
      comment_author = comment['authorDisplayName']
      comment_published_at = comment['publishedAt']
      comments.append((comment_author, comment_text, comment_published_at))
  nextPageToken = response.get('nextPageToken')

  if not nextPageToken:
    break

# Step 5: Filter comments
print("Filter comment...")
hyperlink_pattern = re.compile(
    r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
)

threshold_ratio = 0.65
relavant_comments = []

for author, comment_text, published_at in comments:
  comment_text = comment_text.lower().strip()
  emojis = emoji.emoji_count(comment_text)
  text_characters = len(re.sub(r'\s', '', comment_text))

  if (any(char.isalnum() for char in comment_text)) and not hyperlink_pattern.search(comment_text):
    if emojis == 0 or (text_characters / (text_characters + emojis)) > threshold_ratio:
      relavant_comments.append((author, comment_text, published_at))

# Print relevant comment
print("50 first relevan comment")
for comment in relevant_comments[:50]:
  print(f"Author: {comment[0]}, Comment: {comment[1]}, Published at = {comment[2]}")
```

```python
# Step 5: Save comment to CSV file
with open('youtube_comments.csv', mode='w', newline='', encoding='utf=8') as file:
  writer = csv.writer(file)

  # write header
  writer.writerow(['Author', 'Comment', 'Published'])

  # write comment data
  for comment in comments:
    writer.writerow([comment[0], comment[1], comment[2]])

print("Comments have been saved to youtube_comments.csv")
```

```python
import requests
import json
import csv

# Step 1: Prepere API Key and Search engine ID
API_KEY = 'AIzaSyBUug4PcM5edTDQUwHY1ldH_Z5vXg2zTOE'
SEARCH_ENGINE_ID = '075a3925d0c4d4cca'

# Step 2: Create function to derive results of search with pagination
def search_google_news(query, language='id', time_line = 'd7', results_total = 10):
    news_results = []
    first_index = 1

    while len(news_results) < results_total:
        url = f"https://www.googleapis.com/customsearch/v1?key={API_KEY}&cx={SEARCH_ENGINE_ID}&q={query}&gl=id&num=10&tbm=nws&hl={language}&sort=date&dateRestrict={time_line}&start={first_index}"
        response = requests.get(url)

        # Check if the request was successfull
        if response.status_code == 200:
            search_results = response.json()
            item = search_results.get('items', [])
            news_results.extend(item)

            # Enhance first index a number of 10 for next page results
            first_index += 10

            # if the results back less than from 10, chease loop
            if len(item) < 10:
                break
        else:
            print("Error:", response.status_code)
            break

    return news_results[:results_total]   # Back only results total is wanted
```

```python
# Step 3: Extract and save news_results into CSV
def save_results_to_csv(news_results, file_name='news_results.csv'):

    # Check if news_results without empty
    if news_results:
        with open(file_name, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)

            # Write header
            writer.writerow(['Title', 'Link', 'Description'])

            # Write news data
            for item in news_results:
                title = item['title']
                link = item['link']
                description = item.get('snippet', 'nothing description available')
                writer.writerow([title, link, description])

        print(f"News results have saved in file {file_name}")
    else:
        print("Nothing news results is saved")

# Step 4 Extract and print news results, then save to csv
def print_and_save_news(query, language='id', time_line='d7', results_total = 10):
    news_results = search_google_news(query, language, time_line, results_total)

    if news_results:
        save_results_to_csv(news_results)   #save results to CSV

        # Print news results
        for i, item in enumerate(news_results):
            title = item['title']
            link = item['link']
            description = item.get('snippet', 'nothing description available')
            print(f"{i+1}.Title: {title}")
            print(f"  Link: {link}")
            print(f"  Description: {description}\n")
    else:
        print("Google news not found by your query")

# Step 5: Function Execution with query and deadline
query = "TOM LEMBONG"   # Call with search query what do you want
time_line = 'd7'       # Limit news in last 7 days
results_total = 10     # Determine the results total that you want it
print_and_save_news(query, time_line=time_line, results_total=results_total)
```
